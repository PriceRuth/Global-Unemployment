```python
#Just a quick look on what country has the highest unemployment rate
df.loc[df['2022'].idxmax(),["country_name", "sex", "age_group", "age_categories", "2022"]]
```

```python
#Top 5 countries with the highest unemployment rate
df1 = df.groupby('country_name')[df.columns[5:14]].mean(numeric_only=True).sort_values(by='2022', ascending=False).head(5)
df1
```

```python
df2 = df1.transpose()
df2
```

# Visualization

```python
df2.plot().legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```

It seems like Djibouti, Palestinian Territories, and South Africa are the top three with the highest unemployment rate. In year 2020, it appears that most countries' unemployment rate is decreasing slowly except South Africa.

```python
df_new = df.groupby('country_name')[df.columns[5:14]].mean(numeric_only=True)
df_new
```

```python
df_new = df_new.transpose().reset_index().rename(columns={'index':'Year'})
df_new

```

Let's see the unemployment rate of some of the first world countries.

```python
plt.plot(df_new['Year'], df_new['Australia'], label='Australia')
plt.plot(df_new['Year'], df_new['United States'], label='United States')
plt.plot(df_new['Year'], df_new['Canada'], label='Canada')
plt.legend()
plt.gca().invert_xaxis()
plt.show()
```

In 2020, Australia, US, and Canada had a huge increase in unemployment rate.
